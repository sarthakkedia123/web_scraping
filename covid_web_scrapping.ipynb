```python
from selenium import *
from selenium import webdriver
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
import pandas as pd
```

```python
driver=webdriver.Chrome(executable_path='C:/Users/Hp/Desktop/New folder/web scrapping/chromedriver')
driver.get('https://www.covid19india.org/')

timeout=1000

try:
    WebDriverWait(driver, timeout).until(EC.visibility_of_element_located((By.CLASS_NAME, "level-vaccinated")))
except TimeoutException:
    driver.quit()
#This is done in order to make the driver wait for few seconds before quitting and in that 10 secs it will do the code written in the try.

#The whole scrapping will be done by going on the website and inspecting its code.

# Cases Current Stats
cases_current_stats=driver.find_element(By.CLASS_NAME, "Level").text;
#we request drivert to find out the element by class name level and convert it into the text
print(cases_current_stats)
print(type(cases_current_stats))
print('*************************************')

#formatting the data..splliting the data and storing it in the list
cs_list=cases_current_stats.split('\n')
print(cs_list)
print('*************************************')

print('Today Confirmed:',cs_list[1])
print('Total Confirmed:',cs_list[2])
print('Today Recovered:',cs_list[6])
print('Total Recovered:',cs_list[7])
print('Today Deceased:',cs_list[9])
print('Total Deceased:',cs_list[10])
print('Active Cases:',cs_list[4])
print('*************************************')

# Vaccination Current Stats
vaccination_current_stats=driver.find_element(By.CLASS_NAME,"level-vaccinated")
vcs=vaccination_current_stats.text.split('\n')[0]
print('Total Vaccine Doses:', vcs)
print('*************************************')

vaccination_alo_prog_bar=driver.find_element(By.CLASS_NAME,"progress-bar")
print('At least one dose:',vaccination_alo_prog_bar.text)
print('*************************************')

vaccination_fv_prog_bar=driver.find_elements_by_class_name("label")
fvl=vaccination_fv_prog_bar[1].text
fvl=fvl.split('(')
fvl=fvl[1].split(')')
print('Fully Vaccinated:',fvl[0])
print('*************************************')

'''tested_current_stats=driver.find_element(By.CLASS_NAME,'header-right')
print('Total Tested:',tested_current_stats.text.split('\n')[1])
print('*************************************')'''

category_element=driver.find_element(By.CLASS_NAME,'Table').text
print(category_element)
print('*************************************')
#currently our category_element i.e. our table is in string format..we will convert it in list for easy data visalistion

category_list=category_element.split('\n')
print(category_list)
print('*************************************')

#we are not interested with data having upper arrows or down arrows so we would be removing them
restart=True
while restart:
    for i in range(len(category_list)):
        if(category_list[i].find(chr(8595))!=-1 or category_list[i].find(chr(8593))!=-1):
            category_list.pop(i)
            restart=True
            break
        restart=False
print(category_list)
print('*************************************')

#here we can see that the first two values are randomly generated or we can say that they are false values..
#we can't remove the first 2 values as it will change the index and will create a trouble

print('no. of rows=',(len(category_list)-2)/7)
print('*************************************')

df=pd.DataFrame({category_list[2]:[category_list[i] for i in range(9,len(category_list),7)],
                category_list[3]:[category_list[i] for i in range(10,len(category_list),7)],
                category_list[4]:[category_list[i] for i in range(11,len(category_list),7)],
                category_list[5]:[category_list[i] for i in range(12,len(category_list),7)],
                category_list[6]:[category_list[i] for i in range(13,len(category_list),7)],
                category_list[7]:[category_list[i] for i in range(14,len(category_list),7)],
                category_list[8]:[category_list[i] for i in range(15,len(category_list),7)]})
print(df)
df.to_csv('covid_data.csv')
driver.quit()
```
